```python
from pyalgotrade import strategy
from pyalgotrade import plotter
from pyalgotrade.tools import yahoofinance
from pyalgotrade.technical import vwap
from pyalgotrade.stratanalyzer import returns, sharpe, drawdown, trades
import numpy as np
%matplotlib inline

class VWAPMomentum(strategy.BacktestingStrategy):
    def __init__(self, feed, instrument, vwapWindowSize, threshold):
        strategy.BacktestingStrategy.__init__(self, feed)
        self.__instrument = instrument
        self.__vwap = vwap.VWAP(feed[instrument], vwapWindowSize)
        self.__threshold = threshold

    def getVWAP(self):
        return self.__vwap

    def onBars(self, bars):
        vwap = self.__vwap[-1]
        if vwap is None:
            return

        shares = self.getBroker().getShares(self.__instrument)
        price = bars[self.__instrument].getClose()
        notional = shares * price

        if price > vwap * (1 + self.__threshold) and notional < 1000000:
            self.marketOrder(self.__instrument, 100)
        elif price < vwap * (1 - self.__threshold) and notional > 0:
            self.marketOrder(self.__instrument, -100)


def main(plot):
    instrument = "aapl"
    vwapWindowSize = 5
    threshold = 0.008

    # Download the bars.
    feed = yahoofinance.build_feed([instrument], 2011, 2012, ".")

    strat = VWAPMomentum(feed, instrument, vwapWindowSize, threshold)
    sharpeRatioAnalyzer = sharpe.SharpeRatio()
    strat.attachAnalyzer(sharpeRatioAnalyzer)
    
    drawDownAnalyzer = drawdown.DrawDown()
    strat.attachAnalyzer(drawDownAnalyzer)
    
    returnsAnalyzer = returns.Returns()
    strat.attachAnalyzer(returnsAnalyzer)
    
    tradesAnalyzer = trades.Trades()
    strat.attachAnalyzer(tradesAnalyzer)
    

    if plot:
        plt = plotter.StrategyPlotter(strat, True, False, True)
        plt.getInstrumentSubplot(instrument).addDataSeries("vwap", strat.getVWAP())

    strat.run()
    print "Sharpe ratio: %.2f" % sharpeRatioAnalyzer.getSharpeRatio(0.05)
    
    print "Max drawdown: %f" % drawDownAnalyzer.getMaxDrawDown()
    
    
    print "Cumulative average return was ", 
    x = returnsAnalyzer.getCumulativeReturns()
    
    print np.mean(x)
    
    
    print "Average returns for each bar were: ", 
    print np.mean(returnsAnalyzer.getReturns())
    
    print "Number of trades was %f" % tradesAnalyzer.getCount()
    print "Number of profitable trades was %f" % tradesAnalyzer.getProfitableCount()
    print "Array with profits/losses for each trade: ", tradesAnalyzer.getAll()
    
    print "Array with commision for each trade:",  tradesAnalyzer.getCommissionsForAllTrades()
    
    print "Final portfolio value: $%.2f" %strat.getResult()
    
    if plot:
        plt.plot()

if __name__ == "__main__":
    main(True)
```
